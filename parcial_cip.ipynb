```python
import pandas as pd
import plotly.express as px

file_path = '/content/drive/MyDrive/Colab Notebooks/Datos.csv'
df = pd.read_csv(file_path, sep=",", header=0, encoding="latin-1")

fig = px.bar(df, x='Nombres', y='Total',
             title="Salario Total por Persona",
             labels={'Nombres': 'Nombre', 'Total': 'Salario Total'},
             color='Total',
             color_continuous_scale='Viridis')

fig.show()
```

### üìä **Salario Total por Persona**

La gr√°fica muestra los **salarios totales** de cada persona en la empresa, donde se visualiza claramente qui√©n recibe m√°s dinero. A medida que avanzamos de izquierda a derecha, los **colores m√°s c√°lidos** indican un salario m√°s alto. üî•

üë®‚Äçüíº **Destacado**: La persona que recibe el salario m√°s alto est√° representada por la barra m√°s **alta** y con un color m√°s **intenso**, lo que resalta su total.

Este an√°lisis permite identificar las diferencias salariales y puede ser √∫til para entender la distribuci√≥n de los salarios dentro de la organizaci√≥n. üìàüí∞
